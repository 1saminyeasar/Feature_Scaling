```python
import pandas as pd
```

```python
df = pd.read_csv('supershops.csv')
df.head()
```

```python
x = df.drop(['Profit'],axis=1) # x = df.drop('Profit',axis=1)
y = df[['Profit']]
```

```python
# MinMax Normalization
```

```python
from sklearn.preprocessing import MinMaxScaler
x1=x.copy()
mm = MinMaxScaler()
mm.fit(x1[['Marketing Spend']])
x1[['Marketing Spend']]=mm.transform(x1[['Marketing Spend']])
x1.head()
```

```python
#fit_transform
x1[['Administration']] = mm.fit_transform(x1[['Administration']])
x1.head()
```

```python
#feature_range
mmr = MinMaxScaler(feature_range=(-1,1))
x1[['Transport']]=mmr.fit_transform(x1[['Transport']])
x1[['Transport']]
```

```python
#missing value imputation by mean
x1['Transport'].fillna(x1['Transport'].mean(), inplace = True)
x1
```

# Min Max Scaler using for loop

```python
x2=x.copy()
x2.columns
```

```python
df['Marketing Spend'].dtypes

def func_min_max(i):
    x2[[i]] = mm.fit_transform(x2[[i]])
    
for i in x2:
    if(x2[i].dtypes=='object'):
        continue
    else:
        func_min_max(i)
        
x2.head()
```

# StandardScalar

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x3=x.copy()
x3[['Marketing Spend','Administration','Transport']]=ss.fit_transform(x3[['Marketing Spend','Administration','Transport']])
```

```python
x3
```

```python
# result will be same for separate and together
x5=x.copy()
x5[['Transport']]=ss.fit_transform(x5[['Transport']])
x5[['Administration']]=ss.fit_transform(x5[['Administration']])
x5[['Marketing Spend']]=ss.fit_transform(x5[['Marketing Spend']])
x5.head()
```

# MaxAbsScalar

```python
from sklearn.preprocessing import MaxAbsScaler
mas = MaxAbsScaler()
x4=x.copy()
x4[['Transport']] = mas.fit_transform(x4[['Transport']])
x4[['Administration']] = mas.fit_transform(x4[['Administration']])
x4[['Marketing Spend']] = mas.fit_transform(x4[['Marketing Spend']])
x4
```

```python
#generates same result
x6=x.copy()
x6[['Marketing Spend','Administration','Transport']]=mas.fit_transform(x6[['Marketing Spend','Administration','Transport']])
x6.head()
```

# Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
x7=x.copy()
x7[['Marketing Spend','Administration','Transport']]=rs.fit_transform(x7[['Marketing Spend','Administration','Transport']])
x7.head()
```

```python
#generates same result for together or separate
x8=x.copy()
x8[['Transport']]=rs.fit_transform(x8[['Transport']])
x8.head()
```
